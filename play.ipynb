```python
with open('data/tiny_shakespeare.txt', 'r') as f:
    text = f.read()
```

```python
data = text[:1000]

import tiktoken
enc = tiktoken.get_encoding('gpt2')
tokens = enc.encode(data)
print(tokens[:24])
```

```python
import torch
buf = torch.tensor(tokens[:24 + 1])
x = buf[:-1].view(4, 6)
y = buf[1:].view(4, 6)
print(x)
print(y)
```
